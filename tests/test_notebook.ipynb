<h2> Test: test_username_legal</h2>
<p> The function this tests is <code>fromUserCheck()</code>, which seeks to remove a row that contains a username which does not follow the twitter rules. This could arise through data corruption elsewhere and is important to check. In the example data above rows 5,6 and 8 will stay.</p>

<h2> Test: test_lang_nan_replace</h2>
<p> The function this tests is <code>langCheck()</code>, which seeks to replace a null value in the user_lang column with default value "en". This is because there may be a reason why the langauge cannot be detected but the tweet itself is not corrupted. English was chosen as it is the most common language setting among twitter users. The function also converts everything to lowercase. In the example data above, row 7 user_lang value will change from NaN to en. </p>

```python
def test_lang_nan_replace(self):
        #a null value present
        self.assertTrue(self.df["user_lang"].isnull().any())
        df = rf.langCheck(self.df)
        # nulls replaced with en
        self.assertFalse(df["user_lang"].isnull().any())
        df.to_csv("tests/out/result_test_lang_nan_replace.csv");
```

<h3> Output: </h3>
<p> The output for this function confirms the hypothesis above. The user_lang value in row 7 was replaced.</p>

```python
result_lang
```

<h2> Test: test_drop_nan</h2>
