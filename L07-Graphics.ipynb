# `with` statement

* see https://docs.python.org/3.5/reference/compound_stmts.html#with
* also https://www.python.org/dev/peps/pep-0343/
* allows common `try...except...finally` patterns to be encapsulated for convenient reuse
* wraps the execution of a block with methods defined by a _context manager_
* **context manager** - object that defines the runtime context to be established when executing a `with` statement
* context manager handles the entry into, and the exit from, the desired runtime context for the execution of the block of code

The next example is based on https://gist.github.com/traeblain/1487795

```python
### Make plot
%matplotlib inline
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy import sqrt
from numpy import meshgrid
from numpy import arange

with plt.xkcd():

    ys = arange(-5, 5, 0.01)

    xs = arange(-8, 8, 0.01)
    x, y = meshgrid(xs, ys)
    eq1 = ((x/7)**2*sqrt(abs(abs(x)-3)/(abs(x)-3))+(y/3)**2*sqrt(abs(y+3/7*sqrt(33))/(y+3/7*sqrt(33)))-1)
    plt.contour(x, y, eq1, [0])

    xs = arange(-4, 4, 0.01)
    x, y = meshgrid(xs, ys)
    eq2 = (abs(x/2)-((3*sqrt(33)-7)/112)*x**2-3+sqrt(1-(abs(abs(x)-2)-1)**2)-y)
    plt.contour(x, y, eq2, [0])

    xs = arange(-1, -0.75, 0.01)
    x, y = meshgrid(xs, ys)
    eq3 = (9*sqrt(abs((abs(x)-1)*(abs(x)-.75))/((1-abs(x))*(abs(x)-.75)))-8*abs(x)-y)
    plt.contour(x, y, eq3, [0])

    xs = arange(0.75, 1, 0.01)
    x, y = meshgrid(xs, ys)
    eq3 = (9*sqrt(abs((abs(x)-1)*(abs(x)-.75))/((1-abs(x))*(abs(x)-.75)))-8*abs(x)-y)
    plt.contour(x, y, eq3, [0])

    xs = arange(-0.75, 0.75, 0.01)
    x, y = meshgrid(xs, ys)
    eq4 = (3*abs(x)+.75*sqrt(abs((abs(x)-.75)*(abs(x)-.5))/((.75-abs(x))*(abs(x)-.5)))-y)
    plt.contour(x, y, eq4, [0])

    xs = arange(-0.5, 0.5, 0.01)
    x, y = meshgrid(xs, ys)
    eq5 = (2.25*sqrt(abs((x-.5)*(x+.5))/((.5-x)*(.5+x)))-y)
    plt.contour(x, y, eq5, [0])

    xs = arange(-3, -1, 0.01)
    x, y = meshgrid(xs, ys)
    eq6 = (6*sqrt(10)/7+(1.5-.5*abs(x))*sqrt(abs(abs(x)-1)/(abs(x)-1))-(6*sqrt(10)/14)*sqrt(4-(abs(x)-1)**2)-y)
    plt.contour(x, y, eq6, [0])

    xs = arange(1, 3, 0.01)
    x, y = meshgrid(xs, ys)
    eq6 = (6*sqrt(10)/7+(1.5-.5*abs(x))*sqrt(abs(abs(x)-1)/(abs(x)-1))-(6*sqrt(10)/14)*sqrt(4-(abs(x)-1)**2)-y)
    plt.contour(x, y, eq6, [0])

    plt.show()
```

# Wigdets

From https://github.com/ipython/ipywidgets

```python
from ipywidgets import interact
```

```python
%matplotlib inline
import matplotlib.pyplot as ppt
```

```python
import networkx as netx
```

```python
# wrap a few graph generation functions so they have the same signature

def random_lobster(n, m, k, p):
    return netx.random_lobster(n, p, p / m)

def powerlaw_cluster(n, m, k, p):
    return netx.powerlaw_cluster_graph(n, m, p)

def erdos_renyi(n, m, k, p):
    return netx.erdos_renyi_graph(n, p)

def newman_watts_strogatz(n, m, k, p):
    return netx.newman_watts_strogatz_graph(n, k, p)

def plot_random_graph(n, m, k, p, generator):
    g = generator(n, m, k, p)
    netx.draw(g)
    ppt.show()
```

```python
interact(plot_random_graph, n=(2,30), m=(1,10), k=(1,10), p=(0.0, 1.0, 0.001),
        generator={'lobster': random_lobster,
                   'power law': powerlaw_cluster,
                   'Newman-Watts-Strogatz': newman_watts_strogatz,
                   'Erds-Rnyi': erdos_renyi,
                   });
```

# Maps

Based on the example from "An introduction to Python Programming for Research. Building and maintaining readable, reliable and efficient scientific software. James Hetherington and Matt Clarkson, University College London, 2016. It uses GeoPy (https://github.com/geopy/geopy) - a geocoding library for Python.

```python
import geopy
```

```python
geocoder=geopy.geocoders.GoogleV3(domain="maps.google.co.uk")
```

```python
geocoder.geocode('St Andrews', exactly_one=False)
```

```python
def geolocate(place):
    return geocoder.geocode(place, exactly_one = False)[0][1]
```

```python
home=geolocate('St Andrews')
```

```python
home
```

```python
import requests
def request_map_at(lat,long, satellite=True, zoom=10, size=(400,400), sensor=False):
    base="http://maps.googleapis.com/maps/api/staticmap?"
    params=dict(
        sensor=str(sensor).lower(),
        zoom=zoom,
        size= str(size[0])+"x"+str(size[1]),
        center = str(lat)+","+str(long),
        style="feature:all|element:labels|visibility:off")
    if satellite:
        params["maptype"]="satellite"
    return requests.get(base,params=params)
```

```python
map_response=request_map_at(56.335054, -2.8063431)
```

```python
url=map_response.url
```

```python
url
```

```python
def map_at(*args, **kwargs):
    return request_map_at(*args, **kwargs).content
```

```python
import IPython
map_png=map_at(*home)
```

```python
IPython.core.display.Image(map_png)
```

```python
map_png=map_at(*home, zoom = 15, size=(600,600))
IPython.core.display.Image(map_png)
```
